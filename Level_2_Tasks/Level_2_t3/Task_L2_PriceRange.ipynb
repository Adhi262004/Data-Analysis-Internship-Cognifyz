```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df_valid = df.dropna(subset=["Price range", "Aggregate rating"])
```

```python
price_stats = (
    df_valid.groupby("Price range")
    .agg(
        Restaurant_Count=('Restaurant ID', 'count'),
        Average_Rating=('Aggregate rating', 'mean')
    )
    .reset_index()
    .sort_values(by="Price range")
)
```

```python
price_stats["Average_Rating"] = price_stats["Average_Rating"].round(2)
```

```python
fig, ax1 = plt.subplots(figsize=(8, 5))

ax1.bar(price_stats["Price range"].astype(str), price_stats["Restaurant_Count"], color="skyblue", label="Restaurant Count")
ax1.set_xlabel("Price Range")
ax1.set_ylabel("Number of Restaurants", color="skyblue")
ax1.tick_params(axis='y', labelcolor='skyblue')
```

```python
ax2 = ax1.twinx()
ax2.plot(price_stats["Price range"].astype(str), price_stats["Average_Rating"], color="darkgreen", marker='o', label="Avg Rating")
ax2.set_ylabel("Average Rating", color="darkgreen")
ax2.tick_params(axis='y', labelcolor='darkgreen')

plt.title("Price Range vs Average Rating")
fig.tight_layout()
plt.savefig("PriceRange_vs_Rating.png")  
plt.show()
```
