```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset .csv')
```

```python
df_valid = df.dropna(subset=["Cuisines", "Aggregate rating"])
```

```python
combo_stats = (
    df_valid.groupby("Cuisines")
    .agg(
        Restaurant_Count=('Cuisines', 'count'),
        Average_Rating=('Aggregate rating', 'mean')
    )
    .reset_index()
)
```

```python
combo_stats["Average_Rating"] = combo_stats["Average_Rating"].round(2)
combo_stats = combo_stats.sort_values(by="Restaurant_Count", ascending=False)
```

```python
top10_combos = combo_stats.head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(top10_combos["Cuisines"], top10_combos["Restaurant_Count"], color="mediumseagreen")
plt.xlabel("Number of Restaurants")
plt.title("Top 10 Cuisine Combinations")
plt.gca().invert_yaxis()
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig("Top_Cuisine_Combinations.png")  # Save the chart
plt.show()
```

```python
print("✅ Top 3 Cuisine Combinations with Average Ratings:")
for i in range(3):
    print(f"{i+1}. {top10_combos.iloc[i]['Cuisines']} — "
          f"{top10_combos.iloc[i]['Restaurant_Count']} restaurants, "
          f"Avg Rating: {top10_combos.iloc[i]['Average_Rating']}")
```
