```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
city_counts = (
    df.groupby("City")
    .agg(Restaurant_Count=('Restaurant ID', 'count'))
    .reset_index()
    .sort_values(by="Restaurant_Count", ascending=False)
)
```

```python
top10_cities = city_counts.head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top10_cities["City"], top10_cities["Restaurant_Count"], color="steelblue")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.title("Top 10 Cities by Number of Restaurants")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.savefig("Top_Cities_By_Restaurants.png")
plt.show()
```
