```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset .csv')
```

```python
df_clean = df.dropna(subset=["City","Aggregate rating"])
```

```python
city_counts = df_clean["City"].value_counts().reset_index()
city_counts.columns = ["City", "Restaurant_Count"]
```

```python
top_city_by_count = city_counts.iloc[0]
print("üèôÔ∏è City with Most Restaurants:")
print(top_city_by_count)
```

```python
city_avg_ratings = (
    df_clean.groupby("City")["Aggregate rating"]
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)
city_avg_ratings.columns = ["City", "Average_Rating"]
```

```python
top_city_by_rating = city_avg_ratings.iloc[0]
print("\nüåü City with Highest Average Rating:")
print(top_city_by_rating)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(city_counts["City"].head(10), city_counts["Restaurant_Count"].head(10), color="coral")
plt.title("Top 10 Cities by Number of Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.savefig("Top_Cities_By_Restaurant_Count.png")  # Save to file
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(city_avg_ratings["City"].head(10), city_avg_ratings["Average_Rating"].head(10), color="teal")
plt.title("Top 10 Cities by Average Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.ylim(0, 5)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.savefig("Top_Cities_By_Rating.png")  # Save to file
plt.show()
```

```python
print(f"\n‚úîÔ∏è City with Most Restaurants: {top_city_by_count['City']} ({top_city_by_count['Restaurant_Count']})")
print(f"‚úîÔ∏è City with Highest Average Rating: {top_city_by_rating['City']} ({top_city_by_rating['Average_Rating']:.2f})")
```
