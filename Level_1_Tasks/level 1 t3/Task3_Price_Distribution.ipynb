```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python

price_counts = df["Price range"].value_counts().sort_index().reset_index()
price_counts.columns = ["Price Range", "Restaurant_Count"]
```

```python
total_restaurants = len(df)
price_counts["Percentage"] = (price_counts["Restaurant_Count"]/total_restaurants * 100).round(2)
```

```python
print("Price Range Distribution:")
print(price_counts)
```

```python
plt.figure(figsize=(8, 6))
plt.bar(price_counts["Price Range"].astype(str), price_counts["Restaurant_Count"], color="Red")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.title("Restaurant Distribution by Price Range")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig("Price_Range_Distribution.png") 
plt.show()
```
